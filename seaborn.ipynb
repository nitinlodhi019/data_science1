```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

# **1.Relational Plot**
*   To see the stastical relation between 2 or more variables
*   Bivariate analyssis




## Scatter plot

```python
tips=sns.load_dataset('tips')
tips
```

```python
#scatter plot-> axes level function
sns.scatterplot(data=tips,x='total_bill',y='tip')
```

```python
#  relplot->figure level-> square shape
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter')
```

```python
# hue-> show difference colummn details
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',hue='sex')
```

```python
# style-> Show one more detail in a single graph with some style
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',hue='sex',style='smoker')
```

```python
# size-> show one more detail in aa single graph with the help of size diffrentiation
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',hue='sex',style='smoker',size='day')
```

```python
# facet plot-> work with figure level only
#              plot multiple graph
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',col='sex',hue='time')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',col='sex',row='day')
```

```python
# col_wrap-> here we can decide ho many column in a row.
#            it's provide better redability
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',col='day',col_wrap=2)
```

# 2.Distribution plot
*   Used for univariate analysis(single column)
*   find out distribution



## Histplot

```python
# Axes level plot
sns.histplot(data=tips,x='total_bill')
```

```python
# Figure level plot
sns.displot(data=tips,x='total_bill',kind='hist')
```

```python
# bins-> with the help we can manage bins in graph\


sns.displot(data=tips,x='total_bill',kind='hist',bins=10)
```

```python
# this is a bivariate graph using histogram
sns.displot(data=tips,x='total_bill',y='tip',kind='hist')
```

```python
#KDE(kernal density estimate) Plot
# for visualizing the distribution of observation in a dataset, producing a continuos density estimate.
```

##kdeplot

```python
# axes level plot
sns.kdeplot(data=tips,x='total_bill')
```

```python
# figure level
sns.displot(data=tips,x='total_bill',kind='kde')
```

```python
# fill-> parameter use for filling the area
sns.displot(data=tips,x='total_bill',kind='kde',fill=True)
```

```python
sns.displot(data=tips,x='total_bill',hue='sex',palette="crest",kind='kde')
```

#3.Matrix plot
*   it's a axes level plots
*   it's work on wide range dataset and numerical datasets




```python
# Heatplot
# plot rectangular data as color encoded matrix
```

```python
glue=sns.load_dataset('glue')
```

```python
glue_df=glue.pivot(index="Model", columns="Task", values="Score")
```

```python
sns.heatmap(glue_df)
```

```python
#annot-> provide value of the boxes
sns.heatmap(glue_df,annot=True)
```

```python
sns.heatmap(glue_df,linewidth=0.5,cmap='crest',annot=True)
```

## Clustermap

```python
# Clustermap
# plot a matrix dataset as a hierarchically clustered heatmap
```

```python
sns.clustermap(glue_df)
```
