```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 2-D Line Plot

```python
price=[48000,59000,69000,41000,45000,50000,60000]
year=[2015,2016,2017,2018,2019,2020,2021]
#(categorical_data,numerical_data)
plt.plot(year,price,color='green')
```

```python
batsman=pd.read_csv('/content/sharma-kohli.csv')
batsman
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
# multiple graph(compare to each other)
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# define our title(for understanding)
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title("Rohit sharma VS Virat Kohli career comparison")
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# define our title(for understanding)
# also change the lines color
plt.plot(batsman['index'],batsman['V Kohli'],color='green')
plt.plot(batsman['index'],batsman['RG Sharma'],color='red')
plt.title("Rohit sharma VS Virat Kohli career comparison")
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# cutomize line styles
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashed',linewidth=3)# dashdot
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='solid',linewidth=2)
plt.title("Rohit sharma VS Virat Kohli career comparison")
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# marker and marker size
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashed',linewidth=3,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='solid',linewidth=2,marker='+')#(.)(>)(<)(o)
plt.title("Rohit sharma VS Virat Kohli career comparison")
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# legend--location (labeling of lines)
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashed',linewidth=3,marker='D',markersize=10,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='solid',linewidth=2,marker='+',label='Rohit')#(.)(>)(<)(o)
plt.title("Rohit sharma VS Virat Kohli career comparison")
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend() # plt.legend(loc='set_position')
```

```python
price=[48000,54000,57000,49000,47000,45000,450000]
year=[2015,2016,2017,2018,2019,2020,2021]
plt.plot(year,price)
```

```python
# use limitting axes
price=[48000,54000,57000,49000,47000,45000,450000]
year=[2015,2016,2017,2018,2019,2020,2021]
plt.plot(year,price)
plt.ylim(0,200000)
plt.xlim(2016,2021)
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashed',linewidth=3,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='solid',linewidth=2,marker='+')#(.)(>)(<)(o)
plt.title("Rohit sharma VS Virat Kohli career comparison")
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.grid()
plt.show()
```

# Bar Chart

```python
children=[10,20,40,10,30]
color=['red','blue','yellow','pink','green']
plt.bar(color,children,color='green')
```

```python
plt.barh(color,children,color='yellow')
```

```python
df=pd.read_csv('/content/batsman_season_record.csv')
df
```

```python
plt.bar(df['batsman'],df['2015'],width=0.5)
```

```python
# overlaping
plt.bar(df['batsman'],df['2015'],width=0.5)
plt.bar(df['batsman'],df['2016'],width=0.5)
plt.bar(df['batsman'],df['2017'],width=0.5)
```

```python
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2)
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='red')
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2)
plt.xticks(np.arange(df.shape[0]),df['batsman'])# for names
```

```python
# xticks
children=[10,20,40,10,30]
color=['red red red red','blue blue blue blue','yellow yellow yellow yellow','pink pink pink','green green green green']
plt.bar(color,children,color='green')
plt.xticks(rotation='vertical') #also pass degree
```

```python
df=pd.read_csv('/content/batsman_season_record.csv')
plt.bar(df['batsman'],df['2015'],label='2015')
plt.bar(df['batsman'],df['2016'],bottom=df['2015'],label='2016')
plt.bar(df['batsman'],df['2017'],bottom=df['2016']+df['2015'],label='2017')
plt.legend()
plt.show()
```
